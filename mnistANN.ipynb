```python
import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
datatrain=pd.read_csv("mnist_train.csv")
datatrain.head(5)
```

```python
datatrain.shape
```

```python
datatrain.info()
```

```python
datatrain.isnull().values.any()
```

```python
trainimage=datatrain.iloc[:,1:784]
trainimage.head(2)
```

```python
trainlabel=datatrain.iloc[:,0]
trainlabel.head(7)
```

```python
datatest=pd.read_csv("mnist_test.csv")
datatest.head(5)
```

```python
testimage=datatest.iloc[:,1:784]
testimage.head(2)
```

```python
testlabel=datatest.iloc[:,0]
testlabel.head(7)
```

```python
print(trainimage.shape)
print(testimage.shape)
print(trainlabel.shape)
print(testlabel.shape)
```

```python
trainimage=(trainimage/255)-.5
testimage=(testimage/255)-.5
```

```python
trainimage.head(3)
```

```python
trainlabel.head(5)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense,Dropout,BatchNormalization
```

```python
model=Sequential()
```

```python
model.add(Dense(64,activation="relu", input_shape=(783,)))
```

```python
model.add(Dense(64,activation="relu"))
```

```python
model.add(Dense(90,activation="relu"))
```

```python
model.add(Dense(10,activation="softmax"))
```

```python
model.compile(optimizer="adam",loss="categorical_crossentropy",metrics=["accuracy"])
```

```python
from keras.utils import to_categorical
model.fit(trainimage,to_categorical(trainlabel),epochs=500,batch_size=32)
```

```python
model.evaluate(testimage,to_categorical(testlabel))
```

```python
prediction=model.predict(testimage[:5])
print(prediction)
print(np.argmax(prediction,axis=1))
print(testlabel[:5])
```
